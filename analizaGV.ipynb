```python
# Basic analysis of the data in vsikatalogi_ready.csv

import pandas as pd

# Naložimo podatke
df = pd.read_csv('vsikatalogi_ready.csv', sep=';')

# Prikažemo prvih nekaj vrstic
print("Prvih 5 vrstic podatkov:")
display(df.head())

# Osnovne informacije o podatkih
print("\nOsnovne informacije o podatkih:")
display(df.info())

# Preverimo manjkajoče vrednosti
print("\nŠtevilo manjkajočih vrednosti po stolpcih:")
display(df.isnull().sum())

# Pretvorimo stolpec 'Date' v datetime, če je mogoče
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Osnovna statistika za cene
print("\nOsnovna statistika za cene:")
display(df['Price (€)'].describe())

# Število unikatnih trgovin
print("\nŠtevilo unikatnih trgovin:")
display(df['Store'].nunique())
print("Trgovine:", df['Store'].unique())

# Najpogostejši izdelki
print("\nNajpogostejših 10 izdelkov:")
display(df['Product Name'].value_counts().head(10))

# Najpogostejše trgovine
print("\nNajpogostejših 10 trgovin:")
display(df['Store'].value_counts().head(20))

# Število izdelkov skozi čas (leto)
print("\nŠtevilo izdelkov po letih:")
display(df['Date'].dt.year.value_counts().sort_index())

# Povprečna cena po trgovinah
print("\nPovprečna cena po trgovinah:")
display(df.groupby('Store')['Price (€)'].mean().sort_values(ascending=False))

# Povprečna cena po letih
print("\nPovprečna cena po letih:")
display(df.groupby(df['Date'].dt.year)['Price (€)'].mean())

# Povprečna cena po mesecih (v vseh letih skupaj)
print("\nPovprečna cena po mesecih (vseh let skupaj):")
display(df.groupby(df['Date'].dt.month)['Price (€)'].mean())

# Število izdelkov brez datuma
print("\nŠtevilo izdelkov brez datuma:")
display(df['Date'].isnull().sum())
```

```python
# Najpogostejše besede v stolpcu 'Product Name' brez nezaželenih besed

from collections import Counter
import re

# Seznam nezaželenih besed
stopwords = {"za", "g", "500", "kg", "v", "pp", "natur", "1kg", "100", "200", "po", "z", "s", "mini", "brez", "200", "xxl", "250", "iz"}

# Združimo vse produktne nazive v eno besedilo
all_names = " ".join(df['Product Name'].astype(str).tolist())

# Razbijemo na besede, odstranimo ločila in pretvorimo v male črke
words = re.findall(r'\b\w+\b', all_names.lower())

# Odstranimo nezaželene besede
filtered_words = [word for word in words if word not in stopwords]

# Preštejemo pogostost vsake besede
word_counts = Counter(filtered_words)

# Prikažemo 20 najpogostejših besed
print("Najpogostejših 20 besed v nazivih izdelkov (brez nezaželenih):")
for word, count in word_counts.most_common(20):
    print(f"{word}: {count}")
```

```python
# Izdelki, ki se pojavijo več kot 10x v stolpcu 'Product Name'

product_counts = df['Product Name'].value_counts()
pogosti_izdelki = product_counts[product_counts > 10]

print("Izdelki, ki se pojavijo več kot 10x:")
for izdelek, count in pogosti_izdelki.items():
    print(f"{izdelek}: {count}")
```

```python
# Analiza spremembe cen skozi leta za izdelke, ki se pojavijo več kot 10x

import matplotlib.pyplot as plt

# Najprej izberemo izdelke, ki se pojavijo več kot 10x
product_counts = df['Product Name'].value_counts()
pogosti_izdelki = product_counts[product_counts > 10].index

# Filtriramo podatke za te izdelke
df_pogosti = df[df['Product Name'].isin(pogosti_izdelki)].copy()

# Dodamo stolpec z letom
df_pogosti['Year'] = df_pogosti['Date'].dt.year

# Povprečna cena po letu za vsak izdelek
pivot = df_pogosti.pivot_table(index='Year', columns='Product Name', values='Price (€)', aggfunc='mean')

# Narišemo graf za nekaj najpogostejših izdelkov
najpogostejsi = product_counts[product_counts > 10].head(5).index  # Prvih 5 najpogostejših
pivot[najpogostejsi].plot(marker='o', figsize=(12,6))
plt.title('Povprečna cena skozi leta za najpogostejše izdelke')
plt.ylabel('Povprečna cena (€)')
plt.xlabel('Leto')
plt.legend(title='Izdelek')
plt.grid(True)
plt.show()

# Izračunamo indeks cen (relativno na prvo leto v podatkih) za vsak izdelek
price_index = pivot.copy()
for col in price_index.columns:
    first_valid = price_index[col].dropna().iloc[0] if not price_index[col].dropna().empty else None
    if first_valid:
        price_index[col] = price_index[col] / first_valid * 100

# Prikaz indeksa cen za najpogostejše izdelke
price_index[najpogostejsi].plot(marker='o', figsize=(12,6))
plt.title('Indeks cen skozi leta za najpogostejše izdelke (prvo leto = 100)')
plt.ylabel('Indeks cene')
plt.xlabel('Leto')
plt.legend(title='Izdelek')
plt.grid(True)
plt.show()

# Povprečna sprememba cene za vse pogoste izdelke
print("Povprečna cena po letih za vse pogoste izdelke:")
display(df_pogosti.groupby('Year')['Price (€)'].mean())

print("\nAli je opazna razlika v cenah skozi leta?")
min_year = df_pogosti['Year'].min()
max_year = df_pogosti['Year'].max()
mean_first = df_pogosti[df_pogosti['Year'] == min_year]['Price (€)'].mean()
mean_last = df_pogosti[df_pogosti['Year'] == max_year]['Price (€)'].mean()
print(f"Povprečna cena v {min_year}: {mean_first:.2f} €")
print(f"Povprečna cena v {max_year}: {mean_last:.2f} €")
if mean_last > mean_first:
    print("Cene so se v povprečju zvišale.")
elif mean_last < mean_first:
    print("Cene so se v povprečju znižale.")
else:
    print("Povprečna cena se ni bistveno spremenila.")
```

```python
# Analiza trgovin skozi leta: katere imajo najnižje cene in v kateri trgovini je cena najmanj zrastla

import matplotlib.pyplot as plt

# Dodamo stolpec z letom
df['Year'] = df['Date'].dt.year

# Povprečna cena po trgovinah in letih
pivot_store = df.pivot_table(index='Year', columns='Store', values='Price (€)', aggfunc='mean')

# Prikaz povprečnih cen po trgovinah skozi leta (graf za 5 največjih trgovin po številu izdelkov)
top_stores = df['Store'].value_counts().head(5).index
pivot_store[top_stores].plot(marker='o', figsize=(12,6))
plt.title('Povprečna cena skozi leta za največje trgovine')
plt.ylabel('Povprečna cena (€)')
plt.xlabel('Leto')
plt.legend(title='Trgovina')
plt.grid(True)
plt.show()

# Povprečna cena po trgovinah (vseh let skupaj)
avg_price_by_store = df.groupby('Store')['Price (€)'].mean().sort_values()
print("Trgovine z najnižjo povprečno ceno (vseh let skupaj):")
display(avg_price_by_store.head(10))

# Izračun rasti cen po trgovinah (relativno na prvo leto v podatkih za vsako trgovino)
growth = {}
for store in pivot_store.columns:
    prices = pivot_store[store].dropna()
    if len(prices) > 1:
        first = prices.iloc[0]
        last = prices.iloc[-1]
        growth[store] = (last - first) / first * 100 if first != 0 else None

# Trgovine z najmanjšo rastjo cen
growth_sorted = sorted([(k, v) for k, v in growth.items() if v is not None], key=lambda x: abs(x[1]))
print("Trgovine z najmanjšo absolutno rastjo cen (v %):")
for store, percent in growth_sorted[:10]:
    print(f"{store}: {percent:.2f} %")

# Trgovine z največjo rastjo cen
print("\nTrgovine z največjo rastjo cen (v %):")
for store, percent in sorted(growth.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"{store}: {percent:.2f} %")

# Trgovine, kjer so cene padle
print("\nTrgovine, kjer so cene padle:")
for store, percent in growth.items():
    if percent < 0:
        print(f"{store}: {percent:.2f} %")
```

```python
# Odstranimo outlierje po trgovinah in letih (po IQR metodi)
def remove_outliers(group):
    q1 = group['Price (€)'].quantile(0.25)
    q3 = group['Price (€)'].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return group[(group['Price (€)'] >= lower) & (group['Price (€)'] <= upper)]

# Uporabimo po trgovinah in letih
df_no_outliers = df.groupby(['Store', 'Year'], group_keys=False).apply(remove_outliers)

print(f"Število vrstic pred čiščenjem: {len(df)}")
print(f"Število vrstic po čiščenju: {len(df_no_outliers)}")

# Sedaj lahko analiziraš naprej z df_no_outliers, npr. povprečne cene po letih/trgovinah:
print("\nPovprečna cena po letih za Spar brez outlierjev:")
spar_no_outliers = df_no_outliers[df_no_outliers['Store'] == 'Spar']
display(spar_no_outliers.groupby('Year')['Price (€)'].mean())
```

```python
import pandas as pd
from rapidfuzz import fuzz, process

# Naloži podatke
df = pd.read_csv('vsikatalogi_ready_no_outliers.csv', sep=';')

# Pripravi seznam unikatnih imen izdelkov
product_names = df['Product Name'].unique()
canonical_names = []
name_to_canonical = {}

# Prag za podobnost (0-100)
SIMILARITY_THRESHOLD = 85

for name in product_names:
    found = False
    for canon in canonical_names:
        if fuzz.token_set_ratio(name, canon) >= SIMILARITY_THRESHOLD:
            name_to_canonical[name] = canon
            found = True
            break
    if not found:
        canonical_names.append(name)
        name_to_canonical[name] = name

# Dodaj nov stolpec s kanoničnim imenom
df['Product Name Grouped'] = df['Product Name'].map(name_to_canonical)

# Najpogostejših 10 izdelkov po skupinah
print("Najpogostejših 10 izdelkov (po fuzzy skupinah):")
display(df['Product Name Grouped'].value_counts().head(30))
```

```python
import pandas as pd
from rapidfuzz import fuzz

# Naloži podatke
df = pd.read_csv('vsikatalogi_ready_no_outliers.csv', sep=';')

# Filtriraj samo top 4 trgovine
top_stores = ['Lidl', 'Mercator', 'Hofer', 'Spar']
df_top = df[df['Store'].isin(top_stores)].copy()

# Pripravi seznam unikatnih imen izdelkov samo za te trgovine
product_names = df_top['Product Name'].unique()
canonical_names = []
name_to_canonical = {}

# Prag za podobnost (0-100)
SIMILARITY_THRESHOLD = 85

for name in product_names:
    found = False
    for canon in canonical_names:
        if fuzz.token_set_ratio(name, canon) >= SIMILARITY_THRESHOLD:
            name_to_canonical[name] = canon
            found = True
            break
    if not found:
        canonical_names.append(name)
        name_to_canonical[name] = name

# Dodaj nov stolpec s kanoničnim imenom samo za top trgovine
df_top['Product Name Grouped'] = df_top['Product Name'].map(name_to_canonical)

# Dodaj še en stolpec z ID skupine (številka skupine)
group_ids = {name: idx for idx, name in enumerate(canonical_names)}
df_top['Product Name Grouped ID'] = df_top['Product Name Grouped'].map(group_ids)

# Prikaži nekaj primerov
print("Primeri združenih imen izdelkov za top trgovine:")
display(df_top[['Product Name', 'Product Name Grouped', 'Store']].head(20))

# Po želji shrani v CSV za nadaljnjo analizo
df_top.to_csv('vsikatalogi_ready_no_outliers_topstores_grouped.csv', sep=';', index=False)
```

```python
import pandas as pd

# Naloži podatke z združenimi imeni za top trgovine
df = pd.read_csv('vsikatalogi_ready_no_outliers_topstores_grouped.csv', sep=';')

# Poišči izdelke, ki se pojavijo v vsaj dveh trgovinah
product_store_counts = df.groupby('Product Name Grouped')['Store'].nunique()
common_products = product_store_counts[product_store_counts >= 2].index

# Filtriraj podatke samo na te izdelke
df_common = df[df['Product Name Grouped'].isin(common_products)].copy()

# Povprečna cena po skupinah in trgovinah (samo za skupne izdelke)
pivot = df_common.pivot_table(index='Product Name Grouped', columns='Store', values='Price (€)', aggfunc='mean')

# Za vsak izdelek poišči trgovino z najnižjo povprečno ceno
cheapest_store = pivot.idxmin(axis=1)
cheapest_count = cheapest_store.value_counts()

print("Kolikokrat je posamezna trgovina najcenejša za skupne izdelke:")
print(cheapest_count)

print("\nDelež najcenejših po trgovinah (%):")
print((cheapest_count / len(pivot) * 100).round(2))

# Povprečna cena po trgovinah za skupne izdelke
print("\nPovprečna cena po trgovinah (samo za skupne izdelke):")
print(pivot.mean().sort_values())

# Prikaz nekaj primerov primerjave cen za skupne izdelke
print("\nPrimerjava cen za nekaj skupnih izdelkov:")
display(pivot.head(10))
```

```python
import matplotlib.pyplot as plt

# Bar plot: Kolikokrat je posamezna trgovina najcenejša za skupne izdelke
cheapest_count.plot(kind='bar', color='cornflowerblue')
plt.title('Število najcenejših izdelkov po trgovinah (skupni izdelki)')
plt.ylabel('Število najcenejših izdelkov')
plt.xlabel('Trgovina')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Horizontal bar plot: Delež najcenejših po trgovinah
((cheapest_count / len(pivot)) * 100).sort_values().plot(kind='barh', color='lightseagreen')
plt.title('Delež najcenejših izdelkov po trgovinah (%)')
plt.xlabel('Delež (%)')
plt.ylabel('Trgovina')
plt.tight_layout()
plt.show()

# Bar plot: Povprečna cena po trgovinah za skupne izdelke
pivot.mean().sort_values().plot(kind='bar', color='orange')
plt.title('Povprečna cena po trgovinah (skupni izdelki)')
plt.ylabel('Povprečna cena (€)')
plt.xlabel('Trgovina')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Naloži podatke brez outlierjev
df = pd.read_csv('vsikatalogi_ready_no_outliers.csv', sep=';')

# Filtriraj samo top 4 trgovine
top_stores = ['Lidl', 'Mercator', 'Hofer', 'Spar']
df_top = df[df['Store'].isin(top_stores)].copy()

# Pretvori datum v datetime in izlušči leto
df_top['Date'] = pd.to_datetime(df_top['Date'], errors='coerce')
df_top['Year'] = df_top['Date'].dt.year

# Povprečna cena po trgovinah (vseh izdelkov, vseh let)
avg_price_by_store = df_top.groupby('Store')['Price (€)'].mean().sort_values()
print("Povprečna cena po trgovinah (vseh izdelkov):")
display(avg_price_by_store)

# Povprečna cena po trgovinah in letih (pivot tabela)
pivot = df_top.pivot_table(index='Year', columns='Store', values='Price (€)', aggfunc='mean')

# Vizualizacija: povprečna cena skozi leta za vsako trgovino
pivot.plot(marker='o', figsize=(14,7))
plt.title('Povprečna cena skozi leta po trgovinah (vsi izdelki)')
plt.ylabel('Povprečna cena (€)')
plt.xlabel('Leto')
plt.legend(title='Trgovina')
plt.grid(True)
plt.tight_layout()
plt.show()

# Izračun rasti cen (inflacije) po trgovinah
growth = {}
for store in pivot.columns:
    prices = pivot[store].dropna()
    if len(prices) > 1:
        first = prices.iloc[0]
        last = prices.iloc[-1]
        growth[store] = (last - first) / first * 100 if first != 0 else None

# Prikaz rasti cen po trgovinah
growth_sorted = sorted([(k, v) for k, v in growth.items() if v is not None], key=lambda x: x[1], reverse=True)
print("Rast povprečne cene po trgovinah (od prvega do zadnjega leta, v %):")
for store, percent in growth_sorted:
    print(f"{store}: {percent:.2f} %")

# Vizualizacija: rast cen (inflacija) po trgovinah
plt.figure(figsize=(10,5))
plt.bar([x[0] for x in growth_sorted], [x[1] for x in growth_sorted], color='tomato')
plt.title('Rast povprečne cene po trgovinah (inflacija, v %)')
plt.ylabel('Rast (%)')
plt.xlabel('Trgovina')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Najprej preberimo datoteko kot navaden tekst, da vidimo format
with open('razmerje_cen.csv', 'r', encoding='utf-8') as f:
    lines = f.readlines()[:10]  # Prvih 10 vrstic
    print("Prvih 10 vrstic datoteke:")
    for i, line in enumerate(lines):
        print(f"{i}: {repr(line)}")

# Poskusimo z različnimi separatorji
try:
    # Poskus 1: tab separator
    inflation_df = pd.read_csv('razmerje_cen.csv', sep='\t', header=None)
    print(f"\nS tab separatorjem: {inflation_df.shape} stolpcev")
    print(inflation_df.head())
except:
    pass

try:
    # Poskus 2: brez separatorja (ves tekst v enem stolpcu)
    inflation_df = pd.read_csv('razmerje_cen.csv', sep=None, engine='python', header=None)
    print(f"\nBrez separatorja: {inflation_df.shape} stolpcev")
    print(inflation_df.head())
except:
    pass

# Ročno razdelimo podatke
data = []
with open('razmerje_cen.csv', 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if line and not line.startswith('('):  # Preskoči header
            # Poiščimo tab ali presledek kot separator
            if '\t' in line:
                parts = line.split('\t', 1)
            elif ' ' in line and line.count(' ') == 1:
                parts = line.split(' ', 1)
            else:
                # Poiščimo zadnji presledek/tab pred številko
                import re
                match = re.match(r'([^\d,]+)\s*([0-9,]+)', line)
                if match:
                    parts = [match.group(1).strip(), match.group(2).strip()]
                else:
                    continue
            
            if len(parts) == 2:
                date_str = parts[0].strip()
                price_str = parts[1].strip().replace(',', '.')
                try:
                    price = float(price_str)
                    data.append([date_str, price])
                except:
                    continue

# Ustvari DataFrame
inflation_df = pd.DataFrame(data, columns=['Date_Month', 'Inflation_Index'])

print(f"\nRočno razdelitev: {len(data)} vrstic")
print("Prvih 10 vrstic:")
print(inflation_df.head(10))

# Če imamo podatke, nadaljujemo z analizo
if len(inflation_df) > 0:
    # Pretvori datum v pravilno obliko (iz '2000M01' v '2000-01-01')
    inflation_df['Date_Month_Clean'] = inflation_df['Date_Month'].str.replace('M', '-') + '-01'
    inflation_df['Date_Month_Clean'] = pd.to_datetime(inflation_df['Date_Month_Clean'], errors='coerce')
    
    # Dodaj leto in mesec za lažje povezovanje
    inflation_df['Year'] = inflation_df['Date_Month_Clean'].dt.year
    inflation_df['Month'] = inflation_df['Date_Month_Clean'].dt.month
    
    # Odstrani vrstice z manjkajočimi vrednostmi
    inflation_df = inflation_df.dropna()
    
    print("\nInflacijski indeks po čiščenju (prvih 10 vrstic):")
    display(inflation_df[['Date_Month', 'Inflation_Index', 'Year', 'Month']].head(10))
    
    # Naloži podatke z združenimi imeni za top trgovine
    df = pd.read_csv('vsikatalogi_ready_no_outliers_topstores_grouped.csv', sep=';')
    
    # Pretvori datum
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    
    # Poveži podatke z inflacijskim indeksom
    df_with_inflation = df.merge(
        inflation_df[['Year', 'Month', 'Inflation_Index']], 
        on=['Year', 'Month'], 
        how='left'
    )
    
    # Izračunaj realno ceno (prilagojeno na inflacijo)
    # Formula: realna_cena = nominalna_cena * (100 / inflation_index)
    df_with_inflation['Real_Price'] = df_with_inflation['Price (€)'] * (100 / df_with_inflation['Inflation_Index'])
    
    print(f"\nŠtevilo vrstic z inflacijskim indeksom: {df_with_inflation['Inflation_Index'].notna().sum()}")
    print(f"Število vrstic brez inflacijskega indeksa: {df_with_inflation['Inflation_Index'].isna().sum()}")
    
    # Analiza trgovin z upoštevanjem inflacije (realne cene)
    
    # Poišči izdelke, ki se pojavijo v vsaj dveh trgovinah
    product_store_counts = df_with_inflation.groupby('Product Name Grouped')['Store'].nunique()
    common_products = product_store_counts[product_store_counts >= 2].index
    
    # Filtriraj podatke samo na te izdelke
    df_common_inflation = df_with_inflation[df_with_inflation['Product Name Grouped'].isin(common_products)].copy()
    
    # Povprečna REALNA cena po skupinah in trgovinah (samo za skupne izdelke)
    pivot_real = df_common_inflation.pivot_table(index='Product Name Grouped', columns='Store', values='Real_Price', aggfunc='mean')
    
    # Povprečna NOMINALNA cena po skupinah in trgovinah (samo za skupne izdelke)
    pivot_nominal = df_common_inflation.pivot_table(index='Product Name Grouped', columns='Store', values='Price (€)', aggfunc='mean')
    
    # Za vsak izdelek poišči trgovino z najnižjo povprečno REALNO ceno
    cheapest_store_real = pivot_real.idxmin(axis=1)
    cheapest_count_real = cheapest_store_real.value_counts()
    
    # Za vsak izdelek poišči trgovino z najnižjo povprečno NOMINALNO ceno
    cheapest_store_nominal = pivot_nominal.idxmin(axis=1)
    cheapest_count_nominal = cheapest_store_nominal.value_counts()
    
    print("PRIMERJAVA: NOMINALNE vs REALNE CENE")
    print("=" * 50)
    
    print("\nKolikokrat je posamezna trgovina najcenejša (NOMINALNE cene):")
    print(cheapest_count_nominal)
    print("\nDelež najcenejših po trgovinah - NOMINALNE (%):")
    print((cheapest_count_nominal / len(pivot_nominal) * 100).round(2))
    
    print("\nKolikokrat je posamezna trgovina najcenejša (REALNE cene, prilagojene inflaciji):")
    print(cheapest_count_real)
    print("\nDelež najcenejših po trgovinah - REALNE (%):")
    print((cheapest_count_real / len(pivot_real) * 100).round(2))
    
    print("\nPovprečna nominalna cena po trgovinah (samo za skupne izdelke):")
    print(pivot_nominal.mean().sort_values())
    
    print("\nPovprečna realna cena po trgovinah (samo za skupne izdelke):")
    print(pivot_real.mean().sort_values())
    
    # Primerjava nominalnih in realnih cen
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Graf 1: Delež najcenejših trgovin (nominalne cene)
    (cheapest_count_nominal / len(pivot_nominal) * 100).plot(kind='bar', ax=ax1, color='lightcoral')
    ax1.set_title('Delež najcenejših izdelkov (nominalne cene)')
    ax1.set_ylabel('Delež (%)')
    ax1.tick_params(axis='x', rotation=45)
    
    # Graf 2: Delež najcenejših trgovin (realne cene)
    (cheapest_count_real / len(pivot_real) * 100).plot(kind='bar', ax=ax2, color='lightseagreen')
    ax2.set_title('Delež najcenejših izdelkov (realne cene)')
    ax2.set_ylabel('Delež (%)')
    ax2.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    # Primerjava povprečnih cen
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Graf 1: Povprečne nominalne cene
    pivot_nominal.mean().sort_values().plot(kind='bar', ax=ax1, color='orange')
    ax1.set_title('Povprečna nominalna cena po trgovinah')
    ax1.set_ylabel('Cena (€)')
    ax1.tick_params(axis='x', rotation=45)
    
    # Graf 2: Povprečne realne cene
    pivot_real.mean().sort_values().plot(kind='bar', ax=ax2, color='steelblue')
    ax2.set_title('Povprečna realna cena po trgovinah (prilagojena inflaciji)')
    ax2.set_ylabel('Realna cena (€)')
    ax2.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()
else:
    print("Napaka pri branju inflacijskih podatkov!")
```
