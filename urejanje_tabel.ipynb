Probajmo odstraniti vse številke, ki bi se lahko navezale na količino.

```python
import pandas as pd
import re

def clean_product_name(name):
    # Remove patterns like '1kg', '10 kg', '500 g', 'cena za kos', 'pakiran', etc., and also "za" and "cena"
    pattern = r'(\d+([.,]\d+)?\s?(kg|g|l|ml|kos|L|ML)\b)|(\bcena\b)|(\bcena za kos\b)|(\bpakiran[ao]?\b)|\bza\b'
    name = re.sub(pattern, '', name, flags=re.IGNORECASE)
    # Remove words with 2 or fewer characters
    name = ' '.join([word for word in name.split() if len(word) > 2])
    # Remove extra spaces left after removal
    name = re.sub(r'\s+', ' ', name)
    return name.strip()

# Example usage on your CSV:
df = pd.read_csv('vse_cene_no_outliers.csv', sep=';')
df['Product Name Cleaned'] = df['Product Name'].apply(clean_product_name)
df.to_csv('vse_cene_cleaned.csv', sep=';', index=False)
```

Zdaj ko smo uredili osnovno tabelo, pa naredimo še enkrat grupiranja podobnih izdelkov.

```python
import pandas as pd
from rapidfuzz import fuzz

# Load data
df = pd.read_csv('vse_cene_cleaned.csv', sep=';')

# Use the cleaned product names for grouping
product_names = df['Product Name Cleaned'].dropna().unique()
canonical_names = []
name_to_canonical = {}

SIMILARITY_THRESHOLD = 80

# Step 1: Group similar names
for name in product_names:
    found = False
    for canon in canonical_names:
        if fuzz.token_set_ratio(name, canon) >= SIMILARITY_THRESHOLD:
            name_to_canonical[name] = canon
            found = True
            break
    if not found:
        canonical_names.append(name)
        name_to_canonical[name] = name

df['Product Name Grouped'] = df['Product Name Cleaned'].map(name_to_canonical)

# Step 2: For each group, find the shortest name
group_to_names = {}
for name, group in name_to_canonical.items():
    group_to_names.setdefault(group, []).append(name)

group_to_shortest = {group: min(names, key=len) for group, names in group_to_names.items()}

# Step 3: Map each row to the shortest name in its group
df['Product Name Grouped'] = df['Product Name Grouped'].map(group_to_shortest)

# Save the grouped data
df.to_csv('vse_cene_grouped.csv', sep=';', index=False)

# Print number of unique groups
num_groups = df['Product Name Grouped'].nunique()
print(f"Število različnih grupiranih skupin: {num_groups}")
```
